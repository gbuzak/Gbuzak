```python
#Importação de bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
from scipy.stats import stats
from scipy.stats import chi2_contingency
from statsmodels.graphics.gofplots import qqplot
import statsmodels.api as sm
from scipy.stats import shapiro
from scipy.stats import wilcoxon
from scipy.optimize import curve_fit
from scipy.stats import mannwhitneyu
from sklearn.cluster import KMeans
import pingouin as pg
from fitter import Fitter
```

## Importando e ajustando o dataset

```python
dados = pd.read_csv('DOMensalEstadoDesde1991.csv', sep=';', skip_blank_lines=False) #importando o csv

dados.rename(columns={'vano':'ano', 'letalidade_violenta': 'indicador_letalidade'}, inplace=True) #renomeando colunas

dados = dados[dados['ano'] < 2019] #limitando o dataset até a data desejada

dados = dados[['ano', 'mes', 'indicador_letalidade', 'indicador_roubo_rua', 'indicador_roubo_veic',
               'registro_ocorrencias', 'total_furtos']] #definindo quais colunas serão utilizadas



# Setando os valores nulos de registros de ocorrências com o valor da mediana
dados['registro_ocorrencias'] = pd.to_numeric(dados['registro_ocorrencias'], errors='coerce')
#dados['registro_ocorrencias'] = dados['registro_ocorrencias'].fillna(dados['registro_ocorrencias'].median())
dados['registro_ocorrencias'] = dados['registro_ocorrencias'].astype(float)
display(dados.head(), dados.tail()) #visualização 
```

```python
dados.info()
```

## Criação de variável indicadora da presença ou não de Intervenção Federal no Estado do RJ

```python
#Entre fevereiro de 2018 e dezembro de 2018
dados['presenca_intervencao'] = ((dados['ano'] == 2018) & (dados['mes'] != 2))
print(dados['presenca_intervencao'].value_counts())
```

## Criação de variável clusterizada ('grau_letalidade') por 'K-means'

```python
#Outra técnica de se clusterizar

# import jenkspy
# breaks = jenkspy.jenks_breaks(dados['indicador_letalidade'], nb_class=3)
# dados['grau_letalidade'] = pd.cut(dados['indicador_letalidade'], bins=breaks, labels=['Baixo', 'Medio', 'Alto'], include_lowest=True)
# dados['grau_letalidade'].head(10)
```

```python
dfkmeans = pd.DataFrame(dados['indicador_letalidade'])
dfkmeans

km = KMeans(n_clusters=3, random_state=42)
new = dfkmeans._get_numeric_data()
km.fit(new)
predict=km.predict(new)

dados['grau_letalidade'] = pd.Series(predict, index=dados.index)
dados['grau_letalidade'].value_counts()
dfkmeans = pd.DataFrame(dados['indicador_letalidade'])

km = KMeans(n_clusters=3, random_state=42)
new = dfkmeans._get_numeric_data()
km.fit(new)
predict=km.predict(new)

dados['grau_letalidade'] = pd.Series(predict, index=dados.index)
dados['grau_letalidade'] = dados['grau_letalidade'].astype(str)
dados['grau_letalidade'].replace({'1':'medio', '2':'baixo', '0':'alto'}, inplace=True)
pd.DataFrame(dados['grau_letalidade'].value_counts())
```

### 4.5 Teste exato de Fisher

```python
print('\033[1m' + 'Figura 5. Teste de independência entre as variáveis qualitativas.')
print('\033[0;0m')
quali = pd.crosstab(dados['grau_letalidade'], dados['presenca_intervencao'])
print(quali)      
print('\n')
print(chi2_contingency(quali))
print('\n' + 'p-value: ')
print(round(chi2_contingency(quali)[1],5))

```

<p> Possuindo como objetivo aprofundar a análise dos dados, foram criadas duas
novas variáveis qualitativas, uma indicando a presença ou não de Intervenção Federal, e
a outra que divide em classes os índices de letalidade, gerando três níveis qualitativos
no número de letalidade observadas. </p>

<p> Foi possível, então, realizar o Teste Exato de Fisher (Figura 5), que testa a
independência entre duas variáveis qualitativas. Foi utilizado este teste, em detrimento
do Teste Qui-quadrado, devido ao baixo número de observações de meses com a
presença de Intervenção Federal. Considerando o nível de confiança de 5%, e o p-valor
encontrado de 0,1960 é possível aceitar a hipótese de independência entre as
variáveis estudadas, e afirmar que as mesmas não são correlacionadas. Contudo, é
importante frisar que, como a Intervenção Federal foi realizada em apenas 11 meses, e portanto, o número de observações é baixo, não é possível afirmar de
forma eficiente se existe real independência entre ambas as variáveis testadas. </p>

### 4.6 Teste de Normalidade

```python
print('\033[1m'+'Figura 6: Gráficos de comparação de quantis.')
stats, p = shapiro(dados['indicador_letalidade'])

qqplot(dados['indicador_letalidade'], line='s')
plt.title('Teste de normalidade da variável indicador de letalidade')
plt.show()
print('Shapiro-Wilk ---- p-valor =',  round(p,4))
```

<p>Considerando-se o nível de significância de 5% (ou 0,05) e todos os valores de
p-valor serem menores do que 0,05 no teste de normalidade de Shapiro-Wilk (Figura 6),
é possível descartar a hipótese de distribuição normal em todas as variáveis estudadas.</p>

<p> Tal fato pode ser observado graficamente através dos histogramas (Figura 2) e dos

base sobre os dados dos índices de violência no Estado do Rio de Janeiro, objeto de
estudo deste trabalho, deseja-se saber se as cinco variáveis quantitativas selecionadas
apresentam comportamentos estatísticos iguais ou diferentes após a instalação da

(Figura 8). </p>

<p> É possível constatar, a partir do teste de hipótese de Wilcoxon e dos boxplots
(Figura 9), que as medianas do índice de letalidade antes e após a Intervenção Federal,
bem como os números totais de furtos, não apresentaram diferença estatística,
demonstrando que a presença das Forças Armadas no Estado do Rio de Janeiro não
contribuiu para a redução dos níveis de mortes violentas.</p>

<p> Ao contrário do esperado pelo Governo, após a intervenção houve aumentos
significativos nos níveis de roubos de veículos e de roubos realizados nas ruas do
Estado, assim como o número do registro de ocorrências oficial, o que indica uma piora
na violência urbana em relação a estes indicadores.</p>

<p> Tais resultados iniciais demonstram a ineficácia do método adotado pelo
Governo na Segurança Pública do Estado do Rio de Janeiro. De todo modo, é
importante ressaltar que o número de meses observados (nove, até o momento) é
insuficiente para se chegar a conclusões definitivas.</p>

# 5 CONCLUSÃO
A partir das análises feitas em conjunto à presença da intervenção ou não, pode
se observar que mesmo durante a intervenção o padrão do comportamento da média da
letalidade se manteve. É corroborada, portanto, a pesquisa do Datafolha em que
aproximadamente 52% da população, apesar de aprovar tal intervenção, não tenham
observado qualquer diferença em pouco mais de um mês após o início de tal
interferência. O índice de letalidade no Estado do Rio de Janeiro tendeu à queda de
2002 a 2013, mas vem numa crescente nos últimos anos, o que coincide com a crise
econômica e política. Após os grandes eventos protagonizados pela capital fluminense o
índice de letalidade voltou a crescer e se manter no padrão de letalidade média apesar da
intervenção. É importante ressaltar que o número de meses observados no estudo é
insuficiente para se chegar a conclusões estatísticas definitivas de uma influência da
intervenção na segurança pública, porém os números nesse período ainda são
preocupantes. Pode se constatar então, que estamos diante não só de uma crise política e
econômica mas de segurança também.
